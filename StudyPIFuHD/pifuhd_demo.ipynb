# Pipeline
- 2 Level PIFu
    - Coarse PIFu : Low Resolution Processing
    - Fine PIFu : High Resolution Processing
- Input
    - Raw Image
    - Front / Back Normal Map Image : from pre-trained model(Image-to-Image translation)
- Output : Output of Fine PIFu
